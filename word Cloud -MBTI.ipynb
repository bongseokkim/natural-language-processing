```python
import requests
def read_img_from_url_and_return_matrix(url):
    response = requests.get(url)

    from PIL import Image 
    from io import BytesIO 

    img = Image.open(BytesIO(response.content))
    img_matrix = np.array(img)

    return img_matrix
```
